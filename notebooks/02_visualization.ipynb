# Data Visualization

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display
```

```python
PROCESSED_DATA_FILENAME = '../data/processed/ufo_alcohol.csv'
FIGURE_FILENAME = '../reports/figures/correlation.png'
```

Load preprocessed data

```python
df = pd.read_csv(PROCESSED_DATA_FILENAME, low_memory=False)
```

Take a peek at the first few rows 

```python
display(df.head())
```

Display some dataset stats

```python
df.describe(include='all').T
```

## Plot the dataset

Plot the number of ufo sightings and alcohol consumption per year in the same graph to check for correlations.

```python
plt.rcParams.update({'font.size': 18})

fig, ax = plt.subplots()
fig.set_size_inches(24, 14, forward=True)

df["ufo_sightings"].plot(ax=ax, style='b-')
df["alcohol_consumption"].plot(ax=ax, style='r-', secondary_y=True)

ax.set_xlabel('Year')
ax.set_xticklabels(df["year"])

ax.legend([ax.get_lines()[0], ax.right_ax.get_lines()[0]], 
           ['UFO Sightings', 'Alcolhol Consumption'], bbox_to_anchor=(1.3, 0.5))

plt.show()
```

Save graph

```python
fig.savefig(FIGURE_FILENAME, dpi=fig.dpi, bbox_inches='tight')
```

##   Conclusion 

There does not seem to be a correlation between UFO sightings and alcohol consumption in the USA up until the middle of 1964. Once the alcohol consumption hit an all-time low, ufo sightings began to increase sharply which also led to an increased alcohol consumption... Coincidence?
