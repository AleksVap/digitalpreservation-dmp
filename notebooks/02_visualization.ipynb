# Data Visualization

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display
```

```python
PROCESSED_DATA_FILENAME = '../data/processed/ufo_alcohol.csv'
FIGURE_FILENAME = '../reports/figures/correlation.png'
```

Load preprocessed data

```python
df = pd.read_csv(PROCESSED_DATA_FILENAME, low_memory=False)
```

Take a peek at the first few rows 

```python
display(df.head())
```

Display some dataset stats

```python
df.describe(include='all').T
```

## Plot the dataset

Plot the number of ufo sightings and alcohol consumption per year in the same graph to check for correlations.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

plt.rcParams.update({'font.size': 18})

fig, ax1 = plt.subplots()
fig.set_size_inches(24, 14, forward=True)

ax1.plot(df["year"], df["ufo_sightings"], 'b-')
ax1.set_xlabel('Year')
ax1.set_ylabel('UFO Sightings', color='b')
ax1.tick_params('y', colors='b')

ax1.xaxis.set_major_locator(ticker.MultipleLocator(2))

for tick in ax1.get_xticklabels():
    tick.set_rotation(45)

ax2 = ax1.twinx()
ax2.plot(df["year"], df["alcohol_consumption"], 'r-')
ax2.set_ylabel('Alcohol Consumption (l/capita)', color='r')
ax2.tick_params('y', colors='r')

plt.show()
```

Save graph

```python
fig.savefig(FIGURE_FILENAME, dpi=fig.dpi, bbox_inches='tight')
```

## 👽🍺  Conclusion 🍺👽

There does not seem to be a correlation between UFO sightings and alcohol consumption in the USA up until 1996. Once the alcohol consumption hit an all-time low, ufo sightings began to increase sharply which also led to an increased alcohol consumption... Coincidence?
